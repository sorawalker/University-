```python
from random import randint
```

```python
def sorted_asc(n):
    lst = random_values(n)
    lst.sort()
    return lst


def sorted_desc(n):
    lst = random_values(n)
    lst.sort(reverse=True)
    return lst


def random_values(n):
    return [randint(0,100) for _ in range(n)]

def almost_sorted_asc(n):
    list_nums = random_values(n)
    swap_bool = True
    while swap_bool:
        swap_bool = False
        for i in range(0, len(list_nums) - 1, 2):
            if list_nums[i] > list_nums[i + 1]:
                list_nums[i], list_nums[i + 1] = list_nums[i + 1], list_nums[i]
                swap_bool = True
    return list_nums
```

```python
print(f'—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ –≤–æ–∑. {sorted_asc(10)}')
print(f'—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ —É–±. {sorted_desc(12)}')
print(f'—Å–ª—É—á–∞–π–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è    {random_values(8)}')
print(f'–ø–æ—á—Ç–∏ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π   {almost_sorted_asc(21)}')
```

bubble_sort


```python
def bubble_sort(lst2):
    lst = lst2.copy()
    swaps = 0
    compares = 0
    n = len(lst)
    for i in range(n-1, 0, -1):
        for j in range(i):
            compares += 1
            if lst[j] > lst[j+1]:
                swaps += 1
                lst[j], lst[j+1] = lst[j+1], lst[j]

    return lst, compares, swaps
```

```python
from random import sample

def gen_rand_list(length):
    return sample(range(0, length*10), length)
```

```python
generated = gen_rand_list(10)
print(generated)
print(bubble_sort(generated))
```

```python
def test_sort(sort_alg, list_generator, n, list_lens):
    results = {}
    for length in list_lens:
        comparisons = []
        swaps = []
        for _ in range(n):
            lst = list_generator(length)
            sorted_lst, n_compares, n_swaps = sort_alg(lst)
            comparisons.append(n_compares)
            swaps.append(n_swaps)
            if sorted_lst != sorted(lst):
                return None

        min_compares = min(comparisons)
        max_compares = max(comparisons)
        avg_compares = sum(comparisons) / len(comparisons)
        min_swaps = min(swaps)
        max_swaps = max(swaps)
        avg_swaps = sum(swaps) / len(swaps)

        results[length] = {
                                "min_comp": min_compares,
                                "avg_comp": avg_compares,
                                "max_comp": max_compares,
                                "min_swaps": min_swaps,
                                "avg_swaps": avg_swaps,
                                "max_swaps": max_swaps
                            }

    return results
```

```python
res = test_sort(bubble_sort, gen_rand_list, 20, [10, 100, 200])
for length in res:
    print(length)
    for d in res[length]:
        print(d, res[length][d])
    print()
```

```python
# –æ–±—ã—á–Ω–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º –∏–∑ –ª–µ–∫—Ü–∏–∏
def selection_sort(lst):
    for fill_slot in range(len(lst)-1, 0, -1):
        pos_of_max = 0
        for location in range(1, fill_slot+1):
            if lst[location] > lst[pos_of_max]:
                pos_of_max = location

        lst[fill_slot], lst[pos_of_max] = lst[pos_of_max], lst[fill_slot]
    return lst

print(selection_sort([1,2,12,22,11,12,11,100,20]))
```

```python
# –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º –∏–∑ –ª–µ–∫—Ü–∏–∏
def stone_bubble_sort(arr):
    n = len(arr)
    sublist_length = int(n**0.5)
    while n > 1:
        k = []
        for i in range(0, n, sublist_length):
            sublist = arr[i:min(n, i+sublist_length)]
            max_index = sublist.index(max(sublist))
            max_value = sublist[max_index]
            k.append(max_value)
        max_index = arr.index(max(k))
        arr[max_index], arr[n - 1] = arr[n - 1], arr[max_index]
        n -= 1
    return arr

sorted_arr = stone_bubble_sort([1,2,12,12,0,5,3,10,18,15,14,13,30])
print(sorted_arr)
```

```python
from random import choice
```

```python


def quick_sort(lst: list, count: list):
    if len(lst) <= 1:
        count[0] += 1
        return lst, count

    separator = choice(lst)
    lesser_lst = []
    equal_lst = []
    greater_lst = []

    for el in lst:
        count[0] += 1
        if el < separator:
            count[0] += 1
            lesser_lst.append(el)
        elif el == separator:
            count[0] += 2
            equal_lst.append(el)
        else:
            count[0] += 3
            greater_lst.append(el)

    tmp_lst1, tmp1 = quick_sort(lesser_lst, count)
    tmp_lst2, tmp2 = quick_sort(greater_lst, count)
    return tmp_lst1 + equal_lst + tmp_lst2, count
```

```python
lst = [1, 10, 11, 2, 12, 122, 44, 34]
lst_sorted, count = quick_sort(lst, [0])
print(lst_sorted, *count)
```

```python
import statistics
```

```python
def modified_qsort(lst):
    if len(lst) <= 1:
        return lst
    else:
        pivot = statistics.median([lst[0], lst[len(lst)//2], lst[-1]])  # –≤—ã–±–æ—Ä –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É—è –º–µ–¥–∏–∞–Ω—É
        print(pivot)
        less = [x for x in lst if x < pivot]
        equal = [x for x in lst if x == pivot]
        greater = [x for x in lst if x > pivot]
        return modified_qsort(less) + equal + modified_qsort(greater)
```

–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ–ª—É—á–µ–Ω–∏—è –∏–∑ –±–æ–ª—å—à–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–ø–∏—Å–∫–æ–≤ –æ–¥–Ω–æ–≥–æ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞.

```python
def merge(inp_list_1, inp_list_2):
    list_1 = inp_list_1.copy()
    list_2 = inp_list_2.copy()
    sorted_list = []
    while list_1 and list_2 :
        min_el_1 = min(list_1)
        min_el_2 = min(list_2)
        if min_el_1 < min_el_2:
            sorted_list.append(min_el_1)
            list_1.remove(min_el_1)
        else:
            sorted_list.append(min_el_2)
            list_2.remove(min_el_2)
    if list_1:
        sorted_list = sorted_list + list_1
    if list_2:
        sorted_list = sorted_list + list_2

    return sorted_list
```

```python
def merge_sorted_lists(lists):
    res = []
    for lst in lists:
        res = merge(res, lst)
    return res
```

```python


lists = [
    [1, 2, 3, 5, 77, 100],
    [0, 1, 2, 56, 70, 88, 110],
    [12, 123, 124],
    [0, 12, 100]
]
print(merge_sorted_lists(lists))
```

```python
from random import randint
import time
from math import floor



def gap_insertion_sort(a_list, start, gap):
    for i in range(start + gap, len(a_list), gap):
        current_value = a_list[i]
        position = i
        while position >= gap and a_list[position - gap] > current_value:
            a_list[position] = a_list[position - gap]
            position = position - gap
        a_list[position] = current_value


def shell_sort_mod(a_list):
    temp = 1
    sublist_count = 2 * floor(len(a_list)/(2**(temp+1)))+1
    count = 0
    while True:
        for start_position in range(sublist_count):
            gap_insertion_sort(a_list, start_position, sublist_count)
        print("After inc. of size", sublist_count, "Lst:", a_list)
        temp += 1
        count += 1
        if sublist_count == 1:
            break
        sublist_count = 2 * floor(len(a_list)/(2**(temp+1)))+1
    return count


def shell_sort(a_list):
    sublist_count = len(a_list)//2
    count = 0
    while sublist_count > 0:
        for start_position in range(sublist_count):
            gap_insertion_sort(a_list, start_position, sublist_count)
        print("After inc. of size", sublist_count, "Lst:", a_list)
        count += 1
        sublist_count //= 2
    return count


l = [randint(1, 1000000) for i in range(10000)]
b = l[::]
print(l)
start_time = time.time()
print(shell_sort_mod(l))
print(time.time() - start_time)
print(b)
start2 = time.time()
print(shell_sort(b))
print(time.time() - start2)
```
